```python
import matplotlib.pyplot as plt

from PIL import Image
from math import gcd,sqrt
```

```python
def common_factor(v:tuple) -> list:
    '''
    The function that takes a tuple that contains two vectors as input and returns common divisors by list.
    '''
    cf_lst = []
    gcd_ = gcd(*v)

    for i in range(1, int(sqrt(gcd_)) + 1):
        if gcd_ % i == 0:
            cf_lst.append(i)
            if i**2 != gcd_:
                cf_lst.append(gcd_ // i)

    cf_lst.sort()

    return cf_lst
```

```python
img = Image.open('sample.png')
cf_lst = common_factor(img.size)
```

```python
x,y = img.size

plt.figure(figsize=(10, 4))
plt.subplot(2,4,1)
plt.imshow(img)
plt.axis('off')
plt.title('Raw Image')

for n, i in enumerate(reversed(cf_lst)):
    small_img=img.resize((x//i,y//i),Image.Resampling.BILINEAR)
    resize_img=small_img.resize(img.size,Image.Resampling.NEAREST)
    plt.subplot(2, 4, n+2)
    plt.imshow(resize_img)
    plt.axis('off')
    plt.title(f'{x//i}X{y//i} Dots')
    if n>5:
        plt.show()
        break
```

```python
######################
### save Dot image ###
######################

img = Image.open('sample.png')
dots = (12,10)
newfile_name = f'{img.filename[:-4]}_{dots[0]}X{dots[1]}.png'

small_img=img.resize(dots,Image.Resampling.BILINEAR)
resize_img=small_img.resize(img.size,Image.Resampling.NEAREST)
resize_img.save(f'{newfile_name}')

plt.imshow(resize_img)
plt.axis('off')
plt.show()
```
